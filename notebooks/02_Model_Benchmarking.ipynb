```python
import pandas as pd
import sys
import os
import numpy as np
from pycaret.classification import *
```

```python
df = pd.read_csv("../data/raw/term-deposit-marketing-2020.csv")
print(df.head())
```

```python
# creat an experiment and call compare_models() function to benchmark models on entire dataset

exp1 = setup(df, target='y', session_id=42) 
best_model = compare_models()
```

```python
# Create df_demographic with only demographic and fincial info removed

df_demographic = df.drop(columns=['day', 'month', 'duration','campaign',])
print(df_demographic.head())
```

```python
# Create an experiment and call compare_models() function to benchmark models on df_demographic

exp2 = setup(df_demographic, 
             target='y', 
             session_id=42, 
             categorical_features=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact'],
             fix_imbalance=True, 
             normalize=True)

best_model = compare_models()
```

```python
# Create df_campaign with only campaign features

df_campaign = df.drop(columns=['age', 'education', 'default', 'balance', 
                               'housing', 'loan', 'job', 
                               'contact', 'marital'
                               ])

print(df_campaign.head())

exp3 = setup(df_campaign, target='y', session_id=42)
best_model = compare_models()
```

### We will choose models based on best recall score 
### Beacuse it measures how many actual y = 1 cases were correctly predicted.
### Predicting customers who will say yes is far more important in this context than 
###     mistakenly including a customer who will eventually say no
